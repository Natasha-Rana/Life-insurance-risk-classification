# **1. Introduction**
---


- **Life insurance** refers to the **legally binding contract** between a **policyholder** and an **insurance company** that provides **financial protection to their family.** The **insurer(insurance company)** **promises** to **offer** the **insurance benefit** in exchange for **regular premiums** **paid by** the **life insured(policyholder).**

- The **‘financial protection’** under life insurance is provided in the form of **life cover**, also known as **sum assured.** **Sum assured** is a **pre-agreed amount** that is **payable** in case of an **untoward incident with the life insured.**

- For the **life insurance contract** to be **enforceable**, the application of the **insured** must **accurately disclose** their **current and past health conditions.** Also, the **insured** needs to **pay** a **single premium or regular premiums** as **chosen** when buying life insurance.

- **Life insurance companies** provide life insurance to their customers by **calculating and assigning a risk class to each individual.** Once the **risk class** is **identified by the company,** it **quotes** the **actual cost of the policy** and **premium amount** that is to be **paid by the policyholder** at **regular intervals.**

- For **life insurance companies, risk classes** are used to **determine** **how likely** the **insurance company** is **liable** to **pay out benefits on the behalf of the insured**, if they **pass away.**

- In terms of **premium costs,** several **factors** are used to **determine which risk class the insured fits into.** Those can include:

    - **Smoking status**
    - **Height**
    - **Weight**
    - **Gender**
    - **Family history**
    - **Age**
    - **Occupation**
    - **Whether you engage in risky hobbies or other potentially dangerous behaviors, such as substance or alcohol abuse.**

- **Risk Classifications and Premium Costs:**
The life insurance risk class assigned to the insured can directly impact what they pay for life insurance premiums. Here's an overview of individual risk classes, in general.

**1) Preferred Plus/Elite:** the **lowest-risk category.** **People** in this risk class are in **excellent health,** are typically **younger,** and have **no other immediate cause for concern.** These **people** can expect to **pay** the **lowest premiums** for life insurance.

**2) Preferred:** preferred class policyholders enjoy **lower premiums** due to **excellent health** **but** may have **some subtle red flags** like **higher cholesterol.**

**3) Standard Plus:** people falling under this category have **above-average health,** but things like **blood pressure or body mass index (BMI) may be outside the ideal range.** Standard Plus class policyholders may **pay more than** someone in the **Preferred or Preferred Plus grouping.**

**4) Standard:** This means **typical risk** for the **policyholder,** and **for life insurers,** it means an **average life expectancy.** The **policyholder** may have **some health issues in their family or their past,** which keeps them out of more preferred risk groups, resulting in **higher premiums.**

**5) Substandard/Rated:** If the insured is classified as a **higher risk than standard,** then they are **subject to various degrees or ratings of substandard,** which each insurer approaches a bit differently. This can be because of **health issues or a risky past.** Their **premiums may be among the highest rates,** typically at the **Standard price plus an additional 25 percent at every step down in the ratings.**

**6) Smoker:** Smokers will **pay significantly more due to increased health risks.** Insurers will ask the insured if they smoke or have smoked in the past several years and may test for the presence of nicotine in routine blood work.

- **How Your Risk Class Affects Your Life Insurance Rate:**
  - The **better your rating,** the **lower your insurance premium** will be. The **“standard” rating** is the **baseline,** as     **most people fall into this class.** So, **rates drop** with **each class above standard,** and **rates increase** with       **each table rating below standard.**

# **2. Problem Statement**
---

- **Haplift** one of the **largest issuers** of **life insurance** in the **USA** is planning to **improve** its **life insurance application process.**

- In a one-click shopping world with on-demand everything, the **life insurance application process** of the company is **antiquated/outdated** as well as a **time-consuming process** for the **customers.**

- **Customers** are **required to provide extensive information** to **identify risk classification** and **eligibility,** including **scheduling medical exams.** This **process** takes an average of **30 days.**

- **The result?** **People** are **turned off.** That’s why **only 40% of U.S. households** own **individual life insurance.** **Haplift** wants to **make** this **process quicker** and **less labour-intensive** for **new and existing customers** to **get** a **quote** while **maintaining privacy boundaries.**

- Our **aim** is to **develop** a **predictive model** that **accurately classifies risk** using a more **automated approach.**

- Our **task** is to **predict** the **"Response" variable** for **each Id** in the **test set.** **"Response"** is an **ordinal measure** of **risk** that has **8 levels.**

- **Determination** of the **risk class** for **each customer** is **crucial** for the **company,** so that it can provide **quotes** to its **prospective customers** in **quick and hassel-free manner.**

- The **risk class** varies from **levels 1-8.** **Each risk class** will **correspond** to a **customized quote** for **each customer.**

### 6.4. Converting target variable.

- Here, we will be **converting target variable with multiple classes(1 to 8) to binary/two class(0 or 1).**
- In order to **simplify our classification problem,** we will **modify our target variable** and **convert** our **multi-class classification problem** to **binary class classification problem.**
- **Originally, our target variable has 8 different levels of risk class(i.e.,1-8 ordinal numbers)** making it **difficult to interpret.** Also, **too many classes might interfere with the performance of the model, leading to average results.**
- Therefore, we will **convert the target variable to two classes: 'Standard risk class': 1, 'Non-Standard risk class': 0.**
- This means we will **divide the target variable into two class of customers,** where:
    - **1** will indicate the **customer's life insurance policy** belongs to the **Standard risk class.**
    - **0** will indicate the **customer's life insurance policy** belongs to the **Non-Standard risk class.**
- Since **most of the people** fall under **Standard risk class,** they are required to **pay standard premium rates.**
- On the other hand, **people not** falling under **Standard risk class,** may **enjoy lower premiums or incur higher premiums.** This **premium amount** will **depend** on the **category of risk class assigned** to the **insured.**
- So, **rates drop** with **each class above standard,** and **rates increase** with **each table rating below standard.** 
- Therefore, people falling under the following **risk classes, enjoy lower premium rates as compared to Standard risk class, due to better health conditions:** **Preferred Plus/Elite, Preferred, Standard Plus.** 
- On the other hand, people falling under the following **risk classes, pay higher premium rates as compared to Standard risk class, due to deteriorated health conditions:** **Smokers and Substandard .** 
- The **new problem statement** would be: 
- **Based on the attributes of customers, will the life insurance policy belong to the Standard risk class or not i.e.yes(1) or no(0).**

- This means that customers belonging to **risk class 6 and risk class 8** are **poplular risk classes** against which the company might **approve their life insurance policies.**

````python
# Creating a function to create a new target variable based on conditions:

def new_target(row):
    if (row['Response']<=7) & (row['Response']>=1):
        val=0
    elif (row['Response']==8):
        val=1
    else:
        val=-1
    return val
````

## Approach 1: Feature Selection using Pearson Correlation

- This technique will give us a list of **highly correlated independent features based on the threshold.** We can later **remove highly correlated features as they are serving the same purpose.**
- **Threshold is set at 0.9,** which means that **features having correlation above 0.9** will be **dropped before building the model.**
- Here, we have **selected 113 features** for our models.

````python
threshold=0.90      # Setting the threshold for correlation of features.
````

````python
# Defining a function to remove highly correlated features.
def correlation(dataset, threshold):
    col_corr = set()    # Creating a set of all the names of highly correlated features.
    corr_matrix = dataset.corr()             # Calculating correlation of the entire dataset and storing it in corr_matrix. 
    for i in range(len(corr_matrix.columns)):  # i is the first feature in the dataset which will be compared with j second feature, so on and so forth..
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold: # finding the absolute coefficent values of columns/features having more than 0.8 correlation.
                colname = corr_matrix.columns[i]   # getting the columns.
                col_corr.add(colname)
    return col_corr          # Returning highly correlated columns.
````

#### Mutual information.
- This technique is used when we have a **discrete target variable** having **different classes or categories.**
- **Mutual information (MI) between two random variables is a non-negative value,** which **measures the dependency between the variables.** It is **equal to zero if and only if two random variables are independent, and higher values mean higher dependency.**
- Here, we have **selected 50 features** which have **higher dependencies.**
